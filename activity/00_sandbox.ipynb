```python
import collections
import json
import pandas as pd

from pprint import pprint
from pyairtable import Table
```

```python
with open('credentials.json', 'r') as fp:
    credentials = json.load(fp)
```

```python
base_id = credentials["base_id"]
api_key = credentials["api_key"]
```

```python
def get_df_from_airtable(table_name="JOURNAL", zero_fill=5):
    table = Table(api_key, base_id, table_name)
    rows = collections.defaultdict(dict)
    for i, dic in enumerate(table.all()):
        row_key = f"row_{str(i).zfill(zero_fill)}"
        rows[row_key] = dic["fields"]
        rows[row_key]["id"] = dic["id"]
        rows[row_key]["created_at"] = dic["createdTime"]
    rows_json = json.dumps(rows)
    df = pd.read_json(rows_json, orient='index')
    return df
```
